```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris = pd.read_csv("Iris.csv") 
iris.head()
```

```python
iris.info()
```

```python
iris["Species"].value_counts()
```

```python
sns.scatterplot(data=iris, x="SepalLengthCm", y="SepalWidthCm")
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, height=5)
```

```python
sns.scatterplot(data=iris, x="SepalLengthCm", y="SepalWidthCm", hue="Species")
plt.show()
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
sns.pairplot(iris.drop(columns=["Id"]), hue="Species", height=3)
```
