```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cereal.csv')
```

```python
df
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['shelf'], y=df['calories'])
plt.title('Boxplot of Calories by Shelf')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['calories'], df['rating'], alpha=0.7)
plt.xlabel('Calories')
plt.ylabel('Rating')
plt.title('Scatterplot of Calories vs. Rating')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df_sorted = df.sort_values(by='calories')
plt.plot(df_sorted['calories'], df_sorted['rating'], marker='o', linestyle='-')
plt.xlabel('Calories')
plt.ylabel('Rating')
plt.title('Line Plot of Calories vs. Rating')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['calories'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.title('Histogram of Calories')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['mfr'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title('Distribution of Manufacturers')
plt.ylabel('')
plt.show()
```
